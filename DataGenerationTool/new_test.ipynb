```python
import random
from typing import List

class Ops:
    def __init__(self, numberOfOperated: int):
        self.innerOps: List['Ops'] = []
        self.numberOfOperated = numberOfOperated

        if self.numberOfOperated == 0:
            characters = ["x", "y", "z", "a", "b", "1", "2", "\\alpha", "\\beta", "\\pi", "e"]
        elif self.numberOfOperated == 1:
            characters = [
                "+{{{}}}", "-{{{}}}", "\\sin{{{}}}", "\\cos{{{}}}", "\\tan{{{}}}", "\\log{{{}}}",
                "\\ln{{{}}}", "\\exp{{{}}}", "\\det{{{}}}", "\\sqrt{{{}}}"
            ]
        elif self.numberOfOperated == 2:
            characters = [
                "{{{}}}+{{{}}}", "{{{}}}-{{{}}}", "{{{}}}\\times{{{}}}", "{{{}}}\\cdot{{{}}}",
                "\\frac{{{}}}{{{}}}", "\\binom{{{}}}{{{}}}", "{{{}}}^{{{}}}", "{{{}}}_{{{}}}"
            ]
        elif self.numberOfOperated == 3:
            characters = [
                # tích phân: \int_{a}^{b} f(x)\,dx
                "\\int_{{{}}}^{{{}}}{{{}}}\\,dx",
                # tổng sigma: \sum_{i=1}^{n} a_i
                "\\sum_{{{}}}^{{{}}}{{{}}}",
                # tích pi: \prod_{i=1}^{n} a_i
                "\\prod_{{{}}}^{{{}}}{{{}}}"
            ]
        else:
            characters = ["x"]

        self.character = random.choice(characters)

    def getInnerOps(self, max_depth: int = 4):
        if self.numberOfOperated <= 0:
            return
        weights = [0.6, 0.15, 0.2, 0.05]  # xác suất chọn 0,1,2,3 ngôi
        for _ in range(self.numberOfOperated):
            if max_depth <= 0:
                child = Ops(0)
            else:
                n = random.choices([0, 1, 2, 3], weights=weights)[0]
                child = Ops(n)
                child.getInnerOps(max_depth - 1)
            self.innerOps.append(child)

    def getLatex(self):
        if self.numberOfOperated == 0:
            return self.character
        return self.character.format(*[child.getLatex() for child in self.innerOps])


# --- Demo ---
if __name__ == "__main__":
    for _ in range(5):
        root = Ops(3)      # thử trực tiếp toán tử 3 ngôi
        root.getInnerOps(max_depth=2)
        print(root.getLatex())
```

```python
import matplotlib.pyplot as plt

plt.rcParams["text.usetex"] = False  # Bật full LaTeX

def latex_to_image(latex_code: str, filename="output.png"):
    fig, ax = plt.subplots()
    ax.axis("off")
    ax.text(0.5, 0.5, f"${latex_code}$", fontsize=25, ha="center", va="center")
    plt.savefig(filename, bbox_inches="tight", pad_inches=0.3, dpi=300)
    plt.close()

# Ví 
    
root = Ops(2)   # bắt đầu bằng toán tử 2 ngôi
root.getInnerOps(max_depth=3)
c = 0
latex_to_image(root.getLatex(),"output.png")
c+=1    
```

```python
import pandas as pd
from typing import List
df = pd.DataFrame({"id":List[int],"file_name":List[str],"Latex":List[str]})
for i in range(2000):
    root = Ops(random.randint(2,3))
    root.getInnerOps(max_depth=3)
    num = random.randint(1,99999999)
    Latex = root.getLatex()
    latex_to_image(Latex,"Data\GenData\{num:08d}.png".format(num = num))
    new_record = pd.DataFrame([{"id": i, "file_name": "Data\GenData\{num:08d}.png".format(num = num),"Latex":Latex}])
    df = pd.concat([df, new_record], ignore_index=True)
```
